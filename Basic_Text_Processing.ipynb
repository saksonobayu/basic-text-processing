# Import Library

```python
import re 
import string
import pandas as pd
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

# Import Dataset

```python
df = pd.read_csv("yelp_labelled.csv")
df.head()
```

```python
df.drop(['1','Unnamed: 2','Unnamed: 3','Unnamed: 4','Unnamed: 5'], axis=1, inplace=True)
```

```python
nameDict={"Wow... Loved this place.":"Review"}
df=df.rename(columns=nameDict)
```

```python
df.head()
```

# Case Folding and Removing

```python
df['Review'] = df['Review'].str.lower()
df.head()
```

```python
df['Review'] = df['Review'].str.replace(r'\d+', '', regex=True)
df.head()
```

```python
df['Review'] = df['Review'].str.replace(r'[!”#$%&’()*+,-./:;<=>?@[\]^_`{|}~]', '', regex=True)
df.head()
```

# Filtering

```python
text = stopwords.words('english')
df['Review'] = df['Review'].apply(lambda x: ' '.join([word for word in x.split()if word not in (text)]))
df.head()
```
